# Market Basket Analysis

## Apriori Algorithm

The Apriori algorithm is he foundation of Market basket Analysis and is used to find items frequently bought together. It is useful for retailers because:
- Both X anf Y products can be placed on the same shelf to increase teh sales of one product based on the purchase of other.
- Promotional discounts can be applied to just one item over both items.
- Advertisements on X can be targeted at buyers of Y.
- X and Y could be combined into a new product.

## Affinity Analysis

Affinity analysis is a data analysis and data mining technique that discovers co-occurence relationships among activities performed by specific individuals or groups.

**Association Rule Mining Market Bsket Analysis is one of th ekey techniques used by large retailers to uncover associations between items.**

It works by looking for combinations of items that occur together frequently in transactions.

**Apriori algorithm is an algorithm for frequent itemset mining and association rule learning over relational databases.**

It proceeds by identifying the rfequent individual items in the database and extending them to larger and larger item sets as long as those item sets appear sufficiently often in the database.

*Support:* This says how popular an itemset is, as measured by the proportion of transactions in which the an itemset appears.

*Confidence:* This says how likely an item Y is purchased when the item X is purchased, expressed as {X -> Y}. This is measured by the proportion of transactions with item X, in which item Y also appears.

*Lift:* This says how likely item Y is purchased when item X is purchased while controlling for how popular item Y is.

```python
# ImportLLibraries

import numpy as np
import pandas as pd
import squarify
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
import PyARMViz

from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
from mlxtend.preprocessing import TransactionEncoder
from matplotlib import style
# from PyARMViz import PyARMViz
from PyARMViz.Rule import generate_rule_from_dict

# Visualisation setup
sns.set()
matplotlib.rcParams['figure.figsize'] = (36,36)
style.use('ggplot')
```

```python
# Import Dataset

basket = pd.read_csv('https://github.com/rajeevratan84/datascienceforbusiness/blob/master/Groceries_dataset.csv?raw=true')

basket
```

# Understanding our Data

We have customer or member numbers, and the dates they purchased each item

```python
basket.info()
```

```python
# Convert Date from object to Datetime
basket['Date'] = pd.to_datetime(basket['Date'])

# Sort values for member 1808 by date
basket[basket['Member_number'] == 1808].sort_values(by='Date')
```

```python
# Get the items purchased for each transaction
transactions = [a[1]['itemDescription'].tolist() for a in list(basket.groupby(['Member_number','Date']))]

print(len(transactions))
transactions
```

# Preparing our Dataset

```python
# We use TransactionEncoder to encode our data
te = TransactionEncoder()
te_ary = te.fit(transactions).transform(transactions)
te_ary
```

```python
# Convert to a pandas DataFrame
transactions = pd.DataFrame(te_ary, columns=te.columns_)
transactions
```

Now we create item purchase counts from our transaction data

```python
# transactions.describe()

# Get the count (count - frequency)
pf = transactions.describe()
f = pf.iloc[0] - pf.iloc[3]
a = f.tolist()
b = list(f.index)

# Make a new DataFrame with the count for every item
item = pd.DataFrame([[a[r], b[r]] for r in range(len(a))], columns=['Count','Item'])

# Sort the DataFrame in descending order of count
item = item.sort_values(['Count'], ascending=False).head(50)

item
```

Now let's visualize the 167 items as a Treemap

```python
fig, ax = plt.subplots()

# Set a Color Scheme
cmap = matplotlib.cm.coolwarm

# Get upper and lower bounds for the color mapping
mini = min(item['Count'])
maxi = max(item['Count'])

# Set out color mapping limits
norm = matplotlib.colors.Normalize(vmin=mini, vmax=maxi)

# Obtain our raw colors
colors = [cmap(norm(value)) for value in item['Count']]

# Create the Treemap plot with squarify
squarify.plot(sizes=item['Count'], label=item['Item'], alpha=0.8, color=colors)
plt.axis('off')
plt.title('Top 50 Frequent Basket Items', fontsize=32)
ttl = ax.title
ttl.set_position([0.5, 1.05])
```

## Let's find our Frequent Itemsets using Apriori

```python
frequent_itemsets = apriori(transactions, min_support=0.001, use_colnames=True, max_len=5)
frequent_itemsets['length'] = frequent_itemsets['itemsets'].apply(lambda x: len(x))
frequent_itemsets.head(50)
```

The above gives us all association rules for basket analysis

```python
b = association_rules(frequent_itemsets, metric='lift', min_threshold=0.001)
b['uni'] = np.nan
b['ant'] = np.nan
b['con'] = np.nan
b['tot'] = 14963
```

## Now let's create some illustrative plots

```python
transactions = [a[1]['itemDescription'].tolist() for a in list(basket.groupby(['Member_number', 'Date']))]

def trans():
  for t in transactions:
    yield t
  
def ant(x):
  cnt = 0
  for t in trans():
    t = set(t)
    if x.intersection(t) == x:
      cnt += 1
  
  return cnt

bb = b.values.tolist()
```

```python
rules_dict = []

for bbb in bb:
  bbb[10] = ant(bbb[0])
  bbb[11] = ant(bbb[1])
  bbb[9] = ant(bbb[0].union(bbb[1]))
  diction = {
    'lhs': tuple(bbb[0]),
    'rhs': tuple(bbb[1]),
    'count_full': bbb[9],
    'count_lhs': bbb[10],
    'count_rhs': bbb[11],
    'num_transactions': bbb[12]
  }
  rules_dict.append(diction)
```

```python
rules = []
for rd in rules_dict:
  rules.append(generate_rule_from_dict(rd))
```
